```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
%matplotlib inline
```

```python
#different scales
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
plt.figure()
plt.subplot(221)
plt.plot(x, y)
plt.yscale('log') #logit, symlog, linear
plt.title('log')
plt.grid(True)
```

```python
#Annotate
plt.figure()
plt.subplot(111)
plt.plot(range(len(loan.LoanAmount)), loan.LoanAmount.sort_index(), 'g^')
plt.annotate("max value", xy=(loan.LoanAmount.idxmax(),loan.LoanAmount.max()), xytext=(300,600), arrowprops=dict(facecolor='black', shrink=0.05))

#Text display
plt.text(30, 500, r'$\mu=100,\ \sigma=15$')
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(loan.Property_Area, loan.LoanAmount, alpha=0.2)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')  #set styles
ax.set(xlim=[-1, 800], xlabel='Prop Area', ylabel='Loan Amt',
       title='Loan Data')  #set axes properties
ax.axvline(np.mean(loan.LoanAmount), ls='--', color='r')  #combine plots
ax.fill_between(loan.LoanAmount, 0, 3, color='green', alpha=0.2, where=loan.LoanAmount > np.mean(loan.LoanAmount))
```

```python
#Legend
line_up, = plt.plot([1,2,3], label='Line 2')
line_down, = plt.plot([3,2,1], label='Line 1')
first_legend = plt.legend(handles=[line_up], loc='upper right')
ax = plt.gca().add_artist(first_legend) #add legend manually to axes
sec_legend = plt.legend(handles=[line_down], loc='lower left')
ax = plt.gca().add_artist(sec_legend)
plt.fill_between([1,2], [3,2], 1, alpha=0.2) #fill between 
plt.legend([line_down, line_up], ['orange', 'blue'], loc='best')
```

```python
#Area plot
days = [1, 2, 3, 4, 5]
raining = [7, 8, 6, 11, 7]
snow =  [8, 1, 7, 8, 13]
plt.stackplot(days, raining, snow,colors =['b', 'y'])
plt.xlabel('Days')
plt.ylabel('No of Hours')
plt.title('Representation of Raining and Snow wrt to Days')
plt.show()
```
