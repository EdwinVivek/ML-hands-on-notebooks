```python
class PostEx(Exception):
    def postmethod(self):
        url = "https://services.odata.org/V4/Northwind/Northwind.svc/Orders"
        task = {"Id" : "1"}
        try:
            resp2 = requests.post(url, json=task)
            if resp2.status_code != 201:
                raise ValueError("Failed {}".format(resp2.status_code))
            return ("Success {}".format(resp2.json())) 
        except ValueError as p:
            return p
    
```

```python
class Vehicle:
    count = 0    #class attribute
    
    def __init__(self, name):
        self._name = name
        self.count = 0
    
    def __str__(self):
        return self._name
    
    def bike(self):
        self.count += 1  
        print(f"instance count {self.count}")   # referring instance attribute
        Vehicle.count += 1  # referring class attribute
        print(f"class count {Vehicle.count}")      
```

```python
v = Vehicle("bike")
v.bike()
```

```python
v2 = Vehicle("bike2")
v2.bike()
```
