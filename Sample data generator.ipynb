```python
import json
import random
import string
from faker import Faker

 

# Initialize a Faker object to generate random data
fake = Faker()

 
# Function to generate a random order
def generate_order(order_id):
    customer_name = fake.name()
    email = fake.email()
    shipping_address = fake.address()
    order_date = fake.date_between(start_date='-1y', end_date='today')
    total_amount = round(random.uniform(10, 200), 2)
    items = []
    for _ in range(random.randint(1, 5)):
        product_id = ''.join(random.choices(string.ascii_uppercase, k=3)) + ''.join(random.choices(string.digits, k=3))
        product_name = fake.word()
        quantity = random.randint(1, 10)
        unit_price = round(random.uniform(5, 50), 2)
        item = {
            "product_id": product_id,
            "product_name": product_name,
            "quantity": quantity,
            "unit_price": unit_price
        }
        items.append(item)
    order = {
        "order_id": order_id,
        "customer_name": customer_name,
        "email": email,
        "shipping_address": shipping_address,
        "order_date": order_date.strftime('%Y-%m-%d'),
        "total_amount": total_amount,
        "items": items
    }
    return order



# Generate 1000 random orders
orders = [generate_order(order_id) for order_id in range(1, 1001)]

# Save the orders to a JSON file
with open('sample_orders.json', 'w') as file:
    json.dump(orders, file, indent=2)

print("Sample orders data has been generated and saved to 'sample_orders.json'.")
```

```python
import json
import random
import string
 
def generate_random_string(length=10):
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for _ in range(length))
 
def generate_random_record():
    return {
        "id": random.randint(1, 1000000),
        "name": generate_random_string(),
        "age": random.randint(18, 99),
        "city": generate_random_string(),
        # Add more fields as needed
    }
 
def generate_json_file(file_path, num_records=1000000):
    data = [generate_random_record() for _ in range(num_records)]
    with open(file_path, 'w') as json_file:
        json.dump(data, json_file, indent=2)
 
# Change the file_path to the desired location for the JSON file
file_path = 'sample_data1M.json'
generate_json_file(file_path)
```
