```python
from langchain.document_loaders import WebBaseLoader
from bs4 import BeautifulSoup

def load_and_clean_webpage(url):
    loader = WebBaseLoader(url)
    docs = loader.load()
    # Optional: remove tags, scripts, etc.
    for doc in docs:
        soup = BeautifulSoup(doc.page_content, 'html.parser')
        doc.page_content = soup.get_text()
    return docs
```

```python
URL = "https://github.com/"
docs = load_and_clean_webpage(URL)
docs
```

```python
from langchain.document_loaders import SeleniumURLLoader

urls = [
    "https://help.boldbi.com/embedded-bi/working-with-data-source/data-connectors/google-ads/"
]
loader = SeleniumURLLoader(urls=urls)
docs = loader.load()
print(docs[0].page_content)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunks = splitter.split_text(docs[0].page_content)
chunks
```
